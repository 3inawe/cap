```python
import pandas as pd
import numpy as np
```

```python
df1  =  pd.read_csv('Datasets/Wednesday-workingHours.pcap_ISCX.csv')
df2  =  pd.read_csv('Datasets/Monday-WorkingHours.pcap_ISCX.csv')
df3  =  pd.read_csv('Datasets/Tuesday-WorkingHours.pcap_ISCX.csv', encoding='utf-8')
df4  =  pd.read_csv('Datasets/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
df5  =  pd.read_csv('Datasets/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv', encoding='cp1252', low_memory=False)
df6  =  pd.read_csv('Datasets/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
df7  =  pd.read_csv('Datasets/Friday-WorkingHours-Morning.pcap_ISCX.csv')
df8  =  pd.read_csv('Datasets/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
```

```python
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8]
# Apply the strip operation to the columns of each dataframe
for df in dataframes:
    df.columns = [col.strip() for col in df.columns]
    #print(df.columns)
```

```python
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8]).reset_index(drop=True)
merged_df
```

```python
for column in merged_df.columns:
    # Check if the column data type is 'object', which usually indicates string data
    if merged_df[column].dtype == 'object':
        print("Column",{column},"contains string values.")
```

```python
Datasets2017 = merged_df.drop(['Flow ID', 'Source IP', 'Destination IP', 'Timestamp'], axis  =1)
```

```python
Datasets2017.replace('Infinity', np.inf, inplace=True)

# Replace inf/-inf with NaN (or handle it as needed)
Datasets2017.replace([np.inf, -np.inf], np.nan, inplace=True)

#fill NaN  drop these rows
Datasets2017.dropna(inplace=True)
```

```python
display(Datasets2017.head(2))
```

```python
Datasets2017.Label.value_counts()
```

```python
benign_rows = Datasets2017[Datasets2017['Label'] == 'BENIGN']

#Save the benign rows to a CSV file
benign_rows.to_csv('benign_rows.csv', index=False)
```

```python
benign_rows_cleaning = pd.read_csv('benign_rows.csv')
benign_rows_cleaning.head()
```

```python
duplicate_rows = benign_rows_cleaning.duplicated()
print("Number of duplicate rows:", duplicate_rows.sum())
benign_rows_cleaning.drop_duplicates(inplace=True)
duplicate_rows = benign_rows_cleaning.duplicated()
print("Number of duplicate rows:", duplicate_rows.sum())
```
